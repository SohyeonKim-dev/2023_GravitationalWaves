```python
%load_ext autoreload
%autoreload 2
from problem import Problem
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1.1

```python
u = 1.
v = 0.
N = 600
a = 0.
b = 1.

def src_func(x):
    return - np.pi**2 * np.sin(np.pi * x)
def sol_func(x):
    return np.sin(np.pi * x) + (v - u) / (b - a) * (x - a) + u

prob1 = Problem(1, N, (a,b), sol_func=sol_func, src_func=src_func, bvs=(u,v))
prob1.time
```

```python
prob1.sol.plot(label='numerical sol')
x = prob1.sol.x
plt.plot(x, sol_func(x), label='exact sol')
plt.legend()
```

```python
prob1.res.plot()
```

```python
prob1.err.plot()
```
